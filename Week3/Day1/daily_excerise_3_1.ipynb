```python
import pandas as pd

df = pd.read_csv('clean_dataset.csv')
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

print(df.columns)
```

```python
print(df.nunique())
```

```python
categorical = ['Gender', 'Married', 'BankCustomer', 'Industry','Ethnicity', 'PriorDefault', 'Employed', 'DriversLicense', 'Citizen', 'ZipCode', ]
```

```python
import seaborn as sns
for feature in categorical:
    sns.barplot(data=df, x=feature, y='Income', hue='Approved')
    plt.xticks(rotation=90)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Mental health Depression disorder Data.csv', low_memory=False)
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
numerical_features = ['Schizophrenia (%)', 'Bipolar disorder (%)', 'Eating disorders (%)', 'Anxiety disorders (%)', 'Drug use disorders (%)', 'Depression (%)', 'Alcohol use disorders (%)']

for feature in numerical_features:
    sns.histplot(df[feature].sample(1000))
    plt.show()
```
