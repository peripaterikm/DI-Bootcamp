```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ===== 1. Data Preparation =====

# Генерация данных: температуры от -5 до 35°C (float), 10 городов × 12 месяцев
temp_data = np.random.uniform(low=-5, high=35, size=(10, 12))

# Имена городов
cities = [f"City_{i+1}" for i in range(10)]

# Месяцы
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Создание DataFrame
df = pd.DataFrame(data=temp_data, index=cities, columns=months)

print("Температурные данные (°C):")
print(df.round(1))

# ===== 2. Data Analysis =====

# Вычисление среднегодовой температуры
df["Annual Avg"] = df.mean(axis=1)

# Нахождение самого тёплого и самого холодного города
hottest_city = df["Annual Avg"].idxmax()
coldest_city = df["Annual Avg"].idxmin()

print(f"\nГород с самой высокой средней температурой: {hottest_city}")
print(f"Город с самой низкой средней температурой: {coldest_city}")

# ===== 3. Data Visualization – Линейный график трендов =====

plt.figure(figsize=(12, 6))

# Построение графика по всем городам
for city in cities:
    plt.plot(months, df.loc[city, months], label=city)

plt.title("Тренды температур по месяцам для 10 городов")
plt.xlabel("Месяц")
plt.ylabel("Температура (°C)")
plt.legend(loc='upper right', fontsize='small')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
